```python
# Part 1
import numpy as np

transition = np.matrix([[0.3,0.4,0.3],
                        [0.2,0.5,0.3],
                        [0.4,0.3,0.3]])

# To go two time-steps, we multiply the transition matrix by itself twice A^2
# Upon doing so, we obtain the index 1 row (suburbs, start location), and the index 0 column (downtown, end location)

# The cell in that location gives the probability for part 1.
two_steps = np.dot(transition, transition)
problem1_p1 = two_steps[1,0]
print(problem1_p1)
```

```python
# Part 2

# To calculate reaching the downtown region in two steps for the first time, we subtract the probability of the event where downtown is reached from the previous step with a self-loop.

# The answer would be the answer from part 1 minus P(S->D->D), found from observing the matrix.

transition = np.matrix([[0.3,0.4,0.3],
                        [0.2,0.5,0.3],
                        [0.4,0.3,0.3]])

two_steps = np.dot(transition, transition)
sub_to_downtown_self_loop = 0.2*0.3

problem1_p2 = two_steps[1,0] - sub_to_downtown_self_loop
problem1_p2
```

```python
# Part 4

# Stationary Distribution
# Solve the eigenvalues/eigenvectors equation for the transition matrix A, and observe the eigenvector received at eigenvalue=1. 

# Then, we divide each entry in the eigenvector by the sum of all entries, to make them all add up to 1. 

import numpy as np

def stationary_dist(transition_matrix: np.matrix):
    # NOTE: Remember to transpose your matrix first
    # because it is πA = π1, which is not the same as Aπ = 1π

    # When transposed, πA = A.T*π (like Xβ in machine learning, being the same as βTX)
    matrix = transition_matrix.T
    eigenvalues, eigenvectors = np.linalg.eig(matrix)

    # np.isclose is good to have, as the eigenvalues might not fully equal to 1, but may be 1.0000000003 for example.
    index = np.isclose(eigenvalues, 1)
    
    # return the eigenvector at that location
    # NOTE: eigenvectors are to be read from top to bottom. the rows in the np.matrix are NOT the eigenvectors.

    eigvec_1 = eigenvectors[:,index]
    print(eigenvalues, eigenvectors)
    print(eigvec_1) # make sure the vertical column values are returned

    # now we normalize the values to be between 0-1, so that they can be expressed as probabilities adding up to 1
    eigvec_1 = np.ravel(eigvec_1)
    eigvec_1 = eigvec_1 / sum(eigvec_1)
    return eigvec_1

problem1_stationary = stationary_dist(transition)
problem1_stationary
```

Make a system of equations to get expected steps to reach from one node to another.

$$ 
    m_{i\rightarrow destination} = 
        \begin{cases} 
            0 & \text{if } i \in destination, \\
            1 + \sum_{j \notin destination} p_{ij} m_{j\rightarrow destination} & \text{if } i \notin destination.
        \end{cases}
$$

The case when i is not the destination, intuitively means that we take inter-mediary steps in between, with the initial node, before reaching the destination. 

And it is basically the expected summation of all of those steps that add up to an estimate of the expected steps to reach the final destination. 

```python
# Part 5
# Expected steps from downtown to suburbs can be modelled as an equation to solve, of a collection of intermediary steps, to nodes not being the destination, + 1 = m(i->destination)

# m(i->destination) represents the expected time to reach the destination state. More particularly, we aim at finding m('sub'->'down')

# m(destination->destination) is just 0, as no steps are needed (one is already at the destination node).

# m(suburbs->downtown) = 1 + p(sub->country)m(country->downtown) + p(sub->sub)m(sub->downtown)

# In the above, as a self-loop is also possible, the second term in the summation is present.

# m(country->downtown) = 1 + p(country->suburbs)m(suburbs->downtown) + p(country->country)m(country->downtown)

"""
From the above, we see that there are some common unknown terms on both LHS and RHS, so they simplify. Additionally, we are finding two unknowns (m(country->downtown), m(suburbs->downtown)) and two equations, hence a system of equations can be made.
"""

import numpy as np

A = np.array([[0.5,-0.3],
                 [-0.3,0.7]])
b = np.array([1,1])
x = np.linalg.solve(A,b)

# The first index of x is m(sub->downtown)
# Second index is m(country->downtown)

# Hence, answer to the expected steps from suburbs -> downtown
problem1_ET = x[0]
print(problem1_ET)
```

The features can be emr, hosp, as previous hospitalization history, and emergency room visits may require doctor's office visits for e.g. consultations or treatment steps. Patients with chronic conditions may need doctor's office visits, same as adldiff. Positional variables, like noreast and midwest might also play a role, as due to distance from the patient's homes, the number of visits can be affected. School education can be linked with physician visits, same as the family income. The employed feature plays a similar role as the faminc feature, and those with health insurance might also be more likely to conduct visits to the physicians. With those reasons, the variables are each relevant as features in determining whether a doctor's office would need visits.

```python
# Part 2
problem3_X = problem3_df[['emr','hosp','exclhlth','poorhlth','numchron','adldiff','noreast','midwest','west','age','male','married','school','faminc','employed','privins','medicaid']]
problem3_y = problem3_df['ofp']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(problem3_X, problem3_y, test_size=0.2)
X_train.count(), X_train.shape
```

```python
# Part 3
import scipy.special

class PoissonRegression(object):
    def __init__(self):
        self.coeffs = None
        self.result = None

    def fit(self,X,Y):
        import numpy as np
        import scipy
        from scipy import optimize
    
        # define the objective/cost/loss function we want to minimise
        def loss(coeffs):
            # The parameter lambda for the given X and the proposed values
            # of the coefficients, here coeff[:-1] represent alpha
            # and coeff[-1] represent beta
            lam = np.exp(np.dot(X,coeffs[:-1])+coeffs[-1])

            # use the Y variable that is available here to define
            # the loss function, return the value of the loss for
            # this Y and for this parameter lam defined above
            likelihood = np.sum(lam - np.dot(Y, np.dot(X,coeffs[:-1])+coeffs[-1]))
            return likelihood

        #Use the loss above together with an optimization method from scipy
        #to find the coefficients of the model
        #this is prepared for you below
        initial_arguments = np.zeros(shape=X.shape[1]+1) # initial guess as 0
        self.result = optimize.minimize(loss, initial_arguments, method='cg')
        self.coeffs = self.result.x

    def predict(self,X):
        #Use the trained model to predict Y
        if (self.coeffs is not None):
            return np.exp(np.dot(X,self.coeffs[:-1])+self.coeffs[-1])
```

```python
problem3_model = PoissonRegression()

problem3_model.fit(X_train, y_train)

print(problem3_model.result)
```

```python
# Part 5
from sklearn.metrics import mean_squared_error, accuracy_score

pred = problem3_model.predict(X_test)
mean_squared_error(y_test, pred)
```
